###**Implementando mais informaçoes no gráfico de pizza**

##**Cálculo de Integral Numérica**

###**Primeiro Passo**

Importando: 

```
from scipy.integrate import quad, dblquad
```

````python
from scipy.integrate import quad, dblquad
````

###**Funções**

|  **FUNÇÕES**  |   **DESCRIÇÕES** |
|:----------:|:---------------:|
|**integrate.quad(func, a, b)**| Calcula a integral de uma função |
| **integrate.dblquad(func, a, b, gfun, hfun)**  | Calcula a integral dupla de uma função |

 
 

####**I**

````python
# x**2

f_x = lambda x: x**2
````

````python
#integrate.quad(func, a, b)
#func: nome da variável correspondente a função que você está trabalhando
#a,b: domínios de integração

quad(f_x, 0, 10)
````

````python
#integrate.quad(func, a, b)
#func: nome da variável correspondente a função que você está trabalhando
#a,b: domínios de integração

quad(f_x, 0, 10)[0]
````

####**II**

````python
# x**2 + y

f_xy = lambda x,y: x*y
````

````python
#integrate.dblquad(func, a, b, gfun, hfun)
#func: nome da variável correspondente a função que você está trabalhando
#a,b: domínios de integração
#gfun, hfun: basicamente outro domínio de integração

dblquad(f_xy, 0,5, lambda y:0, lambda y:10)
````

````python
dblquad(f_xy, 0,5, lambda y:0, lambda y:10)[0]
````
